```python
## sales and profit by region

region_data = df.groupby('Region')[['Sales','Profit']].sum().reset_index()
fig, ax = plt.subplots(1, 2, figsize=(15,6))
## Barplot for sales

sns.barplot(data=region_data, x='Region', y='Sales', ax=ax[0], palette='viridis', hue='Region', legend=False)
ax[0].set_title('Total Sales by Region')

## Barplot for Profit

sns.barplot(data=region_data, x='Region', y='Profit', ax=ax[1], palette='viridis', hue='Region', legend=False)
ax[1].set_title('Total Profit by Region')

```

```python
# Bar plot of average quantity and profit by sub-category
sub_cat_analysis = df.groupby('Sub-Category').agg({
    'Quantity': 'mean',
    'Profit': 'mean',
    'Sales': 'mean'
}).round(2)

# Plot 1: Average Quantity by Sub-Category
plt.figure(figsize=(12, 6))
sub_cat_analysis['Quantity'].sort_values().plot(kind='barh', color='skyblue')
plt.title('Average Quantity by Sub-Category')
plt.xlabel('Average Quantity')
plt.tight_layout()
plt.show()

# Plot 2: Average Profit by Sub-Category
plt.figure(figsize=(12, 6))
sub_cat_analysis['Profit'].sort_values().plot(kind='barh', color='lightgreen')
plt.title('Average Profit by Sub-Category')
plt.xlabel('Average Profit')

plt.show()
```

```python
subcategory_data = (
    df.groupby('Sub-Category')
    .agg({'Profit': 'sum', 'Quantity': 'sum'})
    .sort_values('Profit', ascending=False)
    .head(10)
    .reset_index()
)

plt.figure(figsize=(12,6))
sns.barplot(data=subcategory_data, x='Profit', y='Sub-Category', palette='YlGn', hue='Sub-Category', legend=False)

plt.title('Top 10 Profitable Sub-Categories')
plt.xlabel('Profit')
plt.ylabel('Sub-Category')
plt.show()
```

```python
## heatmap of Sales by state and category

plt.figure(figsize=(12, 10))
top_10_states = df.groupby('State')['Sales'].sum().nlargest(10)
state_cat_top10 = df[df['State'].isin(top_10_states.index)].groupby(['State', 'Category'])['Sales'].sum().unstack()
state_cat_top10.plot(kind='barh', stacked=True)
plt.title('Top 10 States - Sales by Category')
plt.xlabel('Sales')
plt.ylabel('State')
plt.legend(title='Category')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Category', y='Quantity')
plt.title('Quantity Distribution by Category')
plt.xticks(rotation=45)
plt.show()
```
